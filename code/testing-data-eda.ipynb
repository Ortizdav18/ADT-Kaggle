```python
# EDA function 
def eda(dataframe):
    print "missing values \n", dataframe.isnull().sum() #count how many null values are in each series
    print "dataframe types \n", dataframe.dtypes # check data types of each colum
    print "dataframe shape \n", dataframe.shape # gives us the number of rows and columns
    print "dataframe describe \n", dataframe.describe() # descriptive statistics
    for item in dataframe:
        print item
        print dataframe[item].nunique() # print number of unique observations 
```

```python
eda(test)
```
